```python
import pandas as pd
```

```python
d1 = pd.read_csv('/cs/labs/oabend/aviramstern/full_model.csv')
d2 = pd.read_csv('/cs/labs/oabend/aviramstern/full_model2.csv')
d = pd.concat([d1, d2])
```

```python
d.sort_values("Tuner Score", ascending=False)
```

```python
def best_in_task(task):
    global d    
    td = d[(d["Task"] == task)].sort_values("Tuner Score", ascending=False)
    best = td.head(1).iloc[0]["Executor ID"]
    print(best)
    return d[
        (d["Executor ID"] == best) & (d["Scope"] == "dev") 
        & d["Class"].isin(['-- All Classes --', "('*', '-- All --')", "('-- All --', '*')"])]
    
```
